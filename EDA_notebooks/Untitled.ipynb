```python
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'lat':[27.11,32.04,41.82,42.36,38.95 ],
                   'long':[114.99,118.76, -71.41,-71.09,-92.33 ],
                   'city':['Jiangxi', 'Nanjing', 'Providence', 'Cambridge', 'Missouri']})

geometry = [Point(xy) for xy in zip(df.long, df.lat)]
gdf = GeoDataFrame(df[['long','lat' ]], geometry=geometry)   

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax=world.plot(color='white', edgecolor='black', figsize=(30,20)), marker='o', color='red', markersize=100, alpha =1);
#plt.savefig('geo_density.png', dpi=300)
```

```python
import geopandas

geometry = [Point(xy) for xy in zip(df.long, df.lat)]
gdf = GeoDataFrame(df[['long','lat' ]], geometry=geometry)   

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax = gdf.plot(ax=world.plot(color='white', edgecolor='blue', figsize=(30,20)), marker='o', color='red', markersize=30, alpha =1);

for x, y, label in zip(df.long, df.lat, df.city):
    ax.annotate(label, xy=(x, y), xytext=(3, 3), textcoords="offset points", fontsize = 5, color='red')
plt.savefig('geo_cities.png', dpi=300)
```
